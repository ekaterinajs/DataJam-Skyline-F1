```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("AQIdata.csv")

month_mapping = {
    "January": 1, "February": 2, "March": 3, "April": 4, "May": 5, "June": 6,
    "July": 7, "August": 8, "September": 9, "October": 10, "November": 11, "December": 12
}
data["month"] = data["Month"].map(month_mapping)

data["date"] = pd.to_datetime(data["Year"].astype(str) + "-" + data["month"].astype(str), format="%Y-%m")

data = data.sort_values("date")

plt.figure(figsize=(12, 6))
sns.lineplot(x="date", y="Median", data=data, marker="o", color="b")

plt.xlabel("Date")
plt.ylabel("Median Value")
plt.title("Trend of Median Values Over Time")
plt.xticks(rotation=45)
plt.grid(True)

plt.show()


plt.figure(figsize=(12, 6))
sns.barplot(x="month", y="Median", hue="Year", data=data, palette="coolwarm")


plt.xlabel("Month")
plt.ylabel("Median Value")
plt.title("Monthly Median Values Across Different Years")
plt.xticks(ticks=range(12), labels=month_mapping.keys(), rotation=45)
plt.legend(title="Year")

plt.show()
```
